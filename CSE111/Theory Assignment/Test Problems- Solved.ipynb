```python
# 1259 - Even and Odd
# Accepted

n = int(input())
even_list = []
odd_list = []


for i in range(n):
    u_input = int(input())
    if u_input % 2 ==0:
        even_list.append(u_input)
    else:
        odd_list.append(u_input)
main_list = sorted(even_list) + sorted(odd_list, reverse=True)
for i in main_list:
    print(i, end='\n')
```

```python
# 2312 - Medal Table
# Accepted

def sort(l):
    for i in range(len(l)):
        maxIndex = i
        for j in range(i + 1, len(l)):
            if l[maxIndex][1] < l[j][1]:
                maxIndex = j
            elif l[maxIndex][1] == l[j][1]:
                if l[maxIndex][2] < l[j][2]:
                    maxIndex = j
                elif l[maxIndex][2] == l[j][2]:
                    if l[maxIndex][0] > l[j][0]:
                        maxIndex = j
        l[maxIndex], l[i] = l[i], l[maxIndex]


def store():
    global n, g
    for i in range(n):
        x = input().split()
        if int(x[1]) not in g:
            g[int(x[1])] = [[x[0], int(x[2]), int(x[3])]]
        else:
            g[int(x[1])].append([x[0], int(x[2]), int(x[3])])
            sort(g[int(x[1])])


def printTable():
    global g
    for k in sorted(g, reverse=True):
        v = g[k]
        for i in range(len(v)):
            print(f"{v[i][0]} {k} {v[i][1]} {v[i][2]}")


n = int(input())
g = {}
store()
printTable()
```

```python
# 2633 - Barbecue at Yuriâ€™s
# Accepted

while True:
    try:
        n = int(input())
        l1 = []
        l2 = []

        for i in range(n):
            x, y = str(input()).split(" ")
            l1.append(x)
            l2.append(int(y))

        for i in range(n):
            min_index = i
            for j in range(i + 1, n):
                if l2[j] < l2[min_index]:
                    min_index = j
            l2[min_index], l2[i] = l2[i], l2[min_index]
            l1[min_index], l1[i] = l1[i], l1[min_index]

        print(*l1, sep=" ")

    except:
        break
```

```python
# 2137-The Library of Mr. Severino
# Accepted

while True:
    try:
        n = int(input())
        list1 = []

        for i in range(n):
            u_input = int(input())
            list1.append(u_input)

        for j in sorted(list1):
            if len(str(j)) == 4:
                print(j)
            elif len(str(j)) == 3:
                print("0" + str(j))
            elif len(str(j)) == 2:
                print("00" + str(j))
            elif len(str(j)) == 1:
                print("000" + str(j))
    except:
        break
```

```python
# 1234 - Dancing Sentence
# Accepted

while True:
    try:
        u_input = str(input())
        out = str()
        count = 0

        for j in u_input:
            if j == " ":
                out += j
            else:
                count += 1
                if count % 2 == 0:
                    out += j.lower()
                else:
                    out += j.upper()

        print(out)

    except:
        break
```

```python
#1238 - Combiner
#Accepted

n = int(input())
new_word = ""
temp = 0

for i in range(n):
    u_input = input()
    x, y = u_input.split(" ")

    if len(x) <= len(y):
        for j in range(len(y)):
            new_word += x[j:j+1]+y[j:j+1]
            temp = j
        new_word += y[temp+1:len(y)]

    else:
        for j in range(len(x)):
            new_word += x[j:j+1]+y[j:j+1]
            temp = j
        new_word += x[temp+1:len(x)]
    print(new_word)
    new_word = ""
    temp = 0
```
