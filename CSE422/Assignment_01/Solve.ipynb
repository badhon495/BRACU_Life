```python
# Gathering heuristic values of all the nodes / manipulating the input text file
input_file = open("Input file.txt", "r")

def heuristicValue(input_file):
    heuristic_own = {}
    heuristic_neighbor = {}
    temp_heuristic_holder = {}

    for each_line in input_file:
        list_of_cities = each_line.strip().split()
        len_of_city_list = len(list_of_cities)

        heuristic_own[list_of_cities[0]] = int(list_of_cities[1]) #own hurestic value

        for city_index in range(2, len_of_city_list, 2): #hurestic distance of connected cities
          temp_heuristic_holder[list_of_cities[city_index]] = int(list_of_cities[city_index+1])

        heuristic_neighbor[list_of_cities[0]] = temp_heuristic_holder
        temp_heuristic_holder = {}

    return (heuristic_own, heuristic_neighbor)

heuristic_own, heuristic_neighbor= heuristicValue(input_file)
# print(heuristic_own)
# print(heuristic_connected)
```

```python
# A start_city algorithm

def aStar(start_city, goal_city, heuristic_neighbor, heuristic_own):
    visited_city_track = []
    heapq.heappush(visited_city_track, (0, start_city))
    actual_path = {}
    cost = {}
    actual_path[start_city] = None
    cost[start_city] = 0


    while visited_city_track:
        current_city = heapq.heappop(visited_city_track)[1]

        if current_city == goal_city:
            break

        for neighbor_city in heuristic_neighbor[current_city]:
            new_cost = cost[current_city] + heuristic_neighbor[current_city][neighbor_city]
            if neighbor_city not in cost or new_cost < cost[neighbor_city]:
                cost[neighbor_city] = new_cost
                priority = new_cost + heuristic_own[neighbor_city]
                heapq.heappush(visited_city_track, (priority, neighbor_city))
                actual_path[neighbor_city] = current_city

    return actual_path, cost
```

```python
# function to generate the required output and backtracking

def outputBuilder(path, cost, start_city, goal_city):
    current_city = goal_city
    actual_path = []

    while current_city is not start_city:
      actual_path.append(current_city)
      new_city = path[current_city]
      current_city = new_city
    actual_path.append(start_city)

    if len(actual_path) != 0:
      actual_path.reverse()
      print(*actual_path, sep = " --> ")
    else:
      print("NO PATH FOUND!")
    print(f'Total distance: {cost[goal_city]} km')
```

```python
#input and call the functions

start_city = input("Enter the start node: ")
goal_city = input("Enter the goal node: ")

if start_city and goal_city not in heuristic_own:
    print("Start or goal does not exist.")
else:
    path, cost = aStar(start_city, goal_city, heuristic_neighbor, heuristic_own)
    outputBuilder(path, cost, start_city, goal_city)
```
