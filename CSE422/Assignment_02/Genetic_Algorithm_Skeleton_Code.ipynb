###Input Processing

```
def inputProcessing(input_file):
    with open(input_file, 'r') as file:
        total_transactions = int(file.readline().strip())
        transaction_types = []
        transaction_amounts = []

        for _ in range(total_transactions):
            transaction = file.readline().strip().split()
            transaction_types.append(transaction[0])
            transaction_amounts.append(int(transaction[1]))

        return total_transactions, transaction_types, transaction_amounts
```

###Chromosome

```
def chromosome(total_transactions):
    chrom_one = ""
    chrom_two = ""

    for _ in range(total_transactions):
        chrom_one += str(rd(0, 1))
        chrom_two += str(rd(0, 1))

    return chrom_one, chrom_two
```

### Fitness function

```
def fitness(chrom_one, chrom_two, transaction_amounts, transaction_types, total_transactions):
    fit_one = 0
    fit_two = 0

    for per_tran in range(total_transactions):
        if chrom_one[per_tran] == "1" and transaction_types[per_tran] == "l":
            fit_one -= transaction_amounts[per_tran]
        if chrom_one[per_tran] == "1" and transaction_types[per_tran] == "d":
            fit_one += transaction_amounts[per_tran]
        if chrom_two[per_tran] == "1" and transaction_types[per_tran] == "l":
            fit_two -= transaction_amounts[per_tran]
        if chrom_two[per_tran] == "1" and transaction_types[per_tran] == "d":
            fit_two += transaction_amounts[per_tran]

    return fit_one, fit_two
```

###Crossover

```
def crossover(chrom_one, chrom_two, total_transactions):
    index = rd(0, total_transactions-1)

    new_chrom_one = chrom_one[:index]+chrom_two[index:]
    new_chrom_two = chrom_two[:index]+chrom_one[index:]

    return new_chrom_one, new_chrom_two
```

###Mutation

```
def mutation(chrom_one, chrom_two, total_transactions):
    index = rd(0, total_transactions-1)

    if chrom_one[index] == "1":
        chrom_one = chrom_one[:index]+"0"+chrom_one[index+1:]
    elif chrom_one[index] == "0":
        chrom_one = chrom_one[:index]+"1"+chrom_one[index+1:]

    if chrom_two[index] == "1":
        chrom_two = chrom_two[:index]+"0"+chrom_two[index+1:]
    elif chrom_two[index] == "0":
        chrom_two = chrom_two[:index]+"1"+chrom_two[index+1:]

    return chrom_one, chrom_two
```

### Genetic Algorithm

```
def geneticAlgorithm(input_file):
    total_transactions, transaction_types, transaction_amounts = inputProcessing(
        input_file)

    for _ in range(100):
        chrom_one, chrom_two = chromosome(total_transactions)
        fit_one, fit_two = fitness(
            chrom_one, chrom_two, transaction_amounts, transaction_types, total_transactions)

        if fit_one == 0 and chrom_one != "0"*total_transactions and len(chrom_one) == total_transactions:
            return chrom_one

        if fit_two == 0 and chrom_two != "0"*total_transactions and len(chrom_two) == total_transactions:
            return chrom_two

        c3, c4 = crossover(chrom_one, chrom_two, total_transactions)
        chrom_three, chrom_four = mutation(c3, c4, total_transactions)
        fit_one, fit_two = fitness(
            chrom_three, chrom_four, transaction_amounts, transaction_types, total_transactions)

        if fit_one == 0 and chrom_three != "0"*total_transactions and len(chrom_three) == total_transactions:
            return chrom_three

        if fit_two == 0 and chrom_four != "0"*total_transactions and len(chrom_four) == total_transactions:
            return chrom_four

    return -1
```
