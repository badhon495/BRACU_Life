```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
import traitlets
class DoublyList:
  def __init__(self, a):
  # Creates a Non Dummy Headed Circular Doubly Linked List using the values from the given array a.
    self.head = Node(a[0], None, None)
    tail = self.head
    for i in range(1, len(a)):
      n = Node(a[i],  None, None)
      tail.next = n
      tail.next.prev = tail
      tail = tail.next
    self.head.prev = tail
    tail.next = self.head

  # Counts the number of Nodes in the list and return the number
  def countNode(self):
    count = 0
    n = self.head.next
    while n != self.head:
        count+=1
        n = n.next
    return count+1
  
  # prints the elements in the list
  def forwardprint(self):
    s = ''
    temp = self.head
    count = self.countNode()
    for i in range(count):
      if temp.next != self.head:
        s+= str(temp.element) + ','
      else:
        s+=str(temp.element) + '.'
      temp = temp.next
    print(s)

  # prints the elements in the list backward
  def backwardprint(self):
    s = ''
    count = self.countNode()
    temp = self.head.prev
    for i in range(count):
      if temp != self.head:
        s+= str(temp.element) + ','
      else:
        s+=str(temp.element) + '.'
      temp = temp.prev
    print(s)

  # returns the reference of the at the given index. For invalid index return None.
  def nodeAt(self, idx):
    count = self.countNode()
    if 0 <= idx < count:
      count = 0
      n = self.head
      while n!= None:
        if count == idx:
          return n
        count+=1
        n = n.next
    else:
      print('index error')
      return None
  
  # returns the index of the containing the given element. if the element does not exist in the List, return -1.
  def indexOf(self, elem):
      temp = self.head
      count = 0
      while temp != None:
          if elem == temp.element:
              return count
          count+=1
          temp = temp.next
      return -1

  # inserts containing the given element at the given index Check validity of index. 
  def insert(self, elem, idx):
    total_nodes = self.countNode()

    if idx == 0:
        new_node = Node(elem, None, None)
        old_head = self.head
        new_node.next = self.head
        new_node.prev = self.head.prev
        self.head.prev.next = new_node
        self.head = new_node
        old_head.prev = self.head

    elif idx >=1 and idx< total_nodes:
      new_node = Node(elem, None, None)
      old_node = self.nodeAt(idx)
      prev_node = old_node.prev

      new_node.next = old_node
      new_node.prev = prev_node
      prev_node.next = new_node
      old_node.prev = new_node

    elif idx == total_nodes:
      new_node = Node(elem, None, None)
      prev_node = self.head.prev
      prev_node.next = new_node
      new_node.prev = prev_node
      new_node.next = self.head
      self.head.prev = new_node
    else:
      print("Invalid Index")
    return self.head.element


  # removes at the given index. returns element of the removed node. Check validity of index. return None if index is invalid.
  def remove(self, idx):
    if idx == 0:
      removed_node = self.head
      self.head = self.head.next
      self.head.prev = removed_node.prev
      removed_node.prev.next = self.head
    elif idx >= 1 and idx < self.countNode():
      n1 = self.nodeAt(idx)
      removed_node = n1
      n1.prev.next = n1.next
      n1.next.prev = n1.prev
    else:
      return None
    return str(removed_node.element)
```
