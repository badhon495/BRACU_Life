```python
class KeyIndex:
    def __init__(self, arr):
        self.size = len(arr)
        self.min = 0
        for i in arr:
            if i < self.min:
                self.min = i
        self.min = -1 * self.min
        max = arr[0]
        for i in arr:
            if i > max:
                max = i
        self.new_arr = [0] * (max + self.min + 1)

        for i in arr:
            self.new_arr[i+self.min] += 1

    def search(self, val):
      if val > len(self.new_arr):
        return False
      elif self.new_arr[val + self.min] == 0:
          return False
      else:
          return True

    def sort(self):
        sort_arr = []
        for i in range(len(self.new_arr)):
            for j in range(self.new_arr[i]):
                sort_arr += [i - self.min]
        return sort_arr


a = [-1, -2, -7, 3, 4, 1, 4, 5, 7, 8,8,8,8,8,8,8,-1, -10]
n = KeyIndex(a)
print(n.search(-7))
print(n.sort())
```

```python
def hashTable(given):
    l1 = [None] * 9
    n = 0
    count = 0
    if len(given) == 0:
        return False
    else:
        for i in given:
            for j in i:
                if ord(j) in range(49, 58):
                    count += int(j)
                elif j not in ["A", "E", "I", "O", "U"]:
                    n += 1

            hash = (n * 24 + count) % 9

            while hash != -1:
                if l1[hash] == None:
                    l1[hash] = i
                    hash = -1
                else:
                    hash = (hash + 1) % 9
            count = 0
            n = 0
        return l1


print(hashTable(['ST1E89B8A32', 'ST1E89B8A32', 'ST1E89B8A32', 'ST1E89B8A32', 'ST1E89B8A32', 'ST1E89B8A32']))
```
