```python
# Run This cell Second
class Node:
  def __init__(self, e, n):
    self.element = e
    self.next = n
```

```python
# To you coding in this cell
class LinkedList:
  def __init__(self, a):
    if type(a) == list:
        self.head = Node(a[0], None)
        tail = self.head
        for i in range(1, len(a)):
          n = Node(a[i],  None)
          tail.next = n
          tail = tail.next
    else:
        self.head = a



  # Traverse elements in the list.
  # This method is done for you. Do not change this method.
  def traverseList(self):
    s = ''
    temp = self.head
    while temp != None:
      if temp.next != None:
        s += str(temp.element) + " "
      else:
        s += str(temp.element)
      temp = temp.next
    return s



  # Count the number of nodes in the list and return the total number
  def countNode(self):
    count = 0
    n = self.head
    while n != None:
        count+=1
        n = n.next
    return count



  # returns the reference of the Node at the given index. For invalid index return None.
  def nodeAt(self, idx):
    if idx < 0 or idx >= self.countNode():
      return None
    else:
      count = 0
      n = self.head
      while n!= None:
        if count == idx:
          return n
        count+=1
        n = n.next
  


  # returns the element of the Node at the given index. For invalid idx return None.
  def get(self, idx):
    if idx < 0 or idx >= self.countNode():
        return None
    else:
        count = 0
        temp = self.head
        while temp != None:
            if count == idx:
                return temp.element
            temp = temp.next
            count += 1



  # updates the element of the Node at the given index. 
  # Returns the old element that was replaced. For invalid index return None.
  # parameter: index, element
  def set(self, idx, elem):
    if idx < 0 or idx >= self.countNode():
        return None
    else:
      count = 0
      temp = self.head
      while temp != None:
          if count == idx:
              rep_value = temp.element
              temp.element = elem
              return rep_value
          temp = temp.next
          count+=1



  # returns the index of the Node containing the given element.
  # if the element does not exist in the List, return -1.
  def indexOf(self, elem):
    temp = self.head
    count = 0
    while temp != None:
        if elem == temp.element:
            return count
        count+=1
        temp = temp.next
    return -1



  # returns true if the element exists in the List, return false otherwise.
  def contains(self, elem):
        temp = self.head
        while temp != None:
            if elem == temp.element:
                return True
            else:
                 temp = temp.next   
        return False
            
    

  # Makes a duplicate copy of the given List. Returns the reference of the duplicate list.
  def copyList(self):
        copy_head = None
        copy_tail = None
        temp = self.head
        while temp != None:
            n = Node(temp.element, None)
            if copy_head == None:
                copy_head = n
                copy_tail = copy_head
            else:
                copy_tail.next = n
                copy_tail = copy_tail.next
            temp = temp.next
        return copy_head



  # Makes a reversed copy of the given List. Returns the head reference of the reversed list.
  def reverseList(self):
        new_head = Node(self.head.element, None)
        temp = self.head.next
        while temp != None:
            n = Node(temp.element, new_head)
            new_head = n
            temp = temp.next
        return new_head



  # inserts Node containing the given element at the given index
  # Check validity of index. If invalid then print "Invalid Index"
  def insert(self, elem, idx):
    total_nodes = self.countNode()
    if idx == 0:
        n = Node(elem, self.head)
        self.head = n
    elif idx >=1 and idx< total_nodes:
        n = Node(elem, self.head)
        n1 = self.nodeAt(idx-1)
        n2 = self.nodeAt(idx)
        n.next = n2
        n1.next = n
    elif idx == total_nodes:
      n = Node(elem, None)
      n1 = self.nodeAt(total_nodes - 1)
      n1.next = n
    else:
      print("Invalid Index")
    return self.head.element



#   # removes Node at the given index. returns element of the removed node.
#   # Check validity of index. return None if index is invalid.
  def remove(self, idx):
    if idx == 0:
      removed_node = self.head
      self.head = self.head.next
    elif idx >= 1 and idx < self.countNode():
      n1 = self.nodeAt(idx-1)
      removed_node = n1.next
      n1.next = removed_node.next
    else:
      return None
    return removed_node.element
  


#   # Rotates the list to the left by 1 position.
  def rotateLeft(self):
    new_head = self.head.next
    temp = new_head
    while temp.next != None:
      temp = temp.next
    temp.next = self.head
    self.head.next = None
    self.head = new_head
    return self.head
  

  
  # Rotates the list to the right by 1 position.
  def rotateRight(self):
    temp = self.head
    while temp.next.next != None:
      temp = temp.next
    old_head = self.head
    self.head = temp.next
    self.head.next = old_head
    temp.next = None
    return self.head
```
