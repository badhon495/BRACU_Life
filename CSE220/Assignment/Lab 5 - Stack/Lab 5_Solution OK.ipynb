```python
# Task 01:
# Array Based Stack:

class Stack:
    def __init__(self):
        self.arr1 = [None]*100
        self.idx = 0
        self.top = None

    def push(self,data):
        if type(data)==int:
            data = str(data)
        
        self.arr1[self.idx] = data
        self.top = data
        self.idx += 1  

    def pop(self):
        if self.idx < 0:  
            return "Stack is empty! Can't pop any element."
        else:
            popped = self.arr1[self.idx-1]
            self.idx -= 1  
            if self.idx>=0:

                self.top = self.arr1[self.idx-1]
                self.arr1[self.idx] = None 
                return popped

    def peek(self):
        if self.idx == 0:
            return "Stack is empty!"
        else:
            return self.top
    
    def parenthesis(self, s1):
        s1 = input("Enter an Expression:")
        idx1 = 1
        for i in s1:
            if self.top == None:
                if i==")" or i=="}" or i=="]":
                    print("This expression is NOT correct.")
                    print("Error at character # "+(str(idx1))+". ‘"+i+"‘- not opened.")
                    return
            if i=="(" or i=="{" or i=="[":
                self.push(idx1)
                self.push(i)
            elif (i==")" and self.top == "(") or (i=="}" and self.top == "{") or (i=="]" and self.top == "["):
 
                self.pop()
 
                self.pop()
            idx1 += 1


        if self.top == None:
            print("This expression is correct.")
            return
        if self.top != None and (self.top == "(" or self.top == "{" or self.top == "["):
            temp = self.top
            self.pop()
            print("This expression is NOT correct.")
            print("Error at character # "+self.top+". ‘"+temp+"‘- not closed.")
            return
        

exp = "1+2*(3/4)"
sk1 = Stack()
sk1.parenthesis(exp)
print()       
```

```python
# Task 02
# Linked List Based Stack:

class Node:
    def __init__(self,data,next):
        self.data = data
        self.next = next

class Stack:
    def __init__(self):
        self.top = None

    def push(self,data):
        if self.top == None: 
            self.top = Node(data,None)
        else:
            n = Node(data,None)
            n.next = self.top      
            self.top = n          
    
    def pop(self):
        if self.top == None:
            return None
        else:
            popped = self.top
            self.top = self.top.next
            popped.next = None
            return popped.data

    def peek(self):
        if self.top == None:
            return None
        else:
            return self.top.data

    def parenthesis(self, s1):
        idx1 = 1
        for i in s1:
            if self.top == None:
                if i==")" or i=="}" or i=="]":
                    print("This expression is NOT correct.")
                    print("Error at character # "+(str(idx1))+". ‘"+i+"‘- not opened.")
                    return
            if i=="(" or i=="{" or i=="[":
                self.push(idx1)
                self.push(i)
            elif (i==")" and self.top.data == "(") or (i=="}" and self.top.data == "{") or (i=="]" and self.top.data == "["):
                self.pop()
                self.pop()
            idx1 += 1
        if self.top == None:
            print("This expression is correct.")
            return
        if self.top != None and (self.top.data == "(" or self.top.data == "{" or self.top.data == "["):
            temp = self.top.data
            self.pop()
            print("This expression is NOT correct.")
            print("Error at character # "+(str(self.top.data))+". ‘"+temp+"‘- not closed.")
            return
        

exp = "1+2*(3/4)"    
sk2 = Stack()
sk2.parenthesis(exp)
print()
```
