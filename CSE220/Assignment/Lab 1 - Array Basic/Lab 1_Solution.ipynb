```python
# Test 03: Shift Right k cell
def shift_right(source, k):
  for i in range(0, k):
        source[k+i] = source[i]
        source[i] = 0
  return source
```

```python
# Test 04: Rotate Right k cell
def rotate_right(source, k):
  for i in range(0, k):
        temp = source[k+i]
        source[k+i] = source[i]
        source[i] = temp
  return source
```

```python
# Test 05: Remove an element from an array
def remove(source, idx):
  for i in range(0, idx+1):
      source[idx+i] = source[idx+i+1]
  return source
```

```python
# Test 06: Remove all occurrences of a particular element from an array
def remove_all(source, element):
    new_array = [0] * len(source)
    idx_count = 0

    for i in source:
        if i != element:
            new_array[idx_count] = i
            idx_count += 1

    return new_array
```

```python
# Test 07: Splitting an Array
def split_array(a):
    balanced = False

    for i in range(len(a)):
        right_side = a[0:i]
        left_side = a[i:]
        right_total = 0
        left_total = 0

        for j in right_side:
            right_total += j
        for k in left_side:
            left_total += k
        if right_total == left_total:
            balanced = True
            break
    return balanced
```
