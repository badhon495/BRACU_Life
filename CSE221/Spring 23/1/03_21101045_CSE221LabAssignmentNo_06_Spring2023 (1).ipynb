```python
#Task 1
input_file = open("/content/sample_data/input1.txt", "r")
nm = input_file.readline().split()
n = int(nm[0])
m = int(nm[1])
adj_list = [[] for _ in range(n)]
for i in range(m):
    edge = input_file.readline().split()
    u = int(edge[0])
    v = int(edge[1])
    w = int(edge[2])
    adj_list[u - 1].append((v - 1, w))

def dijkstra(adj_list, start):
    n = len(adj_list)
    dist = [-1] * n
    dist[start] = 0
    pq = [(0, start)]
    while pq:
        d, u = min(pq)
        pq.remove((d, u))
        if d != dist[u]:
            continue
        for v, w in adj_list[u]:
            if dist[v] == -1 or dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
                pq.append((dist[v], v))
    return dist


start_node = int(input_file.readline())
dist = dijkstra(adj_list, start_node - 1)

output_file = open("/content/sample_data/output1", mode="w")
for d in dist:
    output_file.write(f"{d} ")
    
input_file.close()
output_file.close() 
```

```python
#Task 2
input_file = open("/content/sample_data/input 2.txt", "r")
n_m = input_file.readline().split()
num_nodes = int(n_m[0])
num_edges = int(n_m[1])

adj_list = [[] for _ in range(num_nodes)]
for i in range(num_edges):
    edge = input_file.readline().split()
    u = int(edge[0]) - 1
    v = int(edge[1]) - 1
    weight = int(edge[2])
    adj_list[u].append((v, weight))

start_nodes = input_file.readline().split()
start_A = int(start_nodes[0]) - 1
start_B = int(start_nodes[1]) - 1

def dijkstra(adj, start):
    n = len(adj)
    dist = [-1] * n
    dist[start] = 0
    pq = [(0, start)]
    while pq:
        d, u = min(pq)
        pq.remove((d, u))
        if d != dist[u]:
            continue
        for v, w in adj[u]:
            if dist[v] == -1 or dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
                pq.append((dist[v], v))
    return dist

dist_A = dijkstra(adj_list, start_A)
dist_B = dijkstra(adj_list, start_B)

meet_node = -1
min_dist = float('inf')
for i in range(num_nodes):
    if dist_A[i] != -1 and dist_B[i] != -1 and dist_A[i] + dist_B[i] < min_dist:
        min_dist = dist_A[i] + dist_B[i]
        meet_node = i

output_file = open("/content/sample_data/output2", "w")
if meet_node == -1:
    output_file.write("Impossible")
else:
    output_file.write(str(min_dist) + " " + str(meet_node+1))
output_file.close()

input_file.close()

if meet_node == -1:
    print("Impossible")
else:
    print("Time", min_dist)
    print("Node", meet_node+1)
```

```python
#Task 3
def dijkstra(graph, start, end):
    n = len(graph)
    distances = [float('inf')] * n
    distances[start] = 0
    visited = [False] * n
    
    for i in range(n):
        
        min_dist = float('inf')
        min_node = -1
        for j in range(n):
            if not visited[j] and distances[j] < min_dist:
                min_dist = distances[j]
                min_node = j
        
        if min_node == -1:
            return -1
        
        visited[min_node] = True
        
        if min_node == end:
            return distances[end]
        
        for neighbor in graph[min_node]:
            neighbor_node = neighbor[0]
            weight = neighbor[1]
            danger = max(distances[min_node], weight)
            if danger < distances[neighbor_node]:
                distances[neighbor_node] = danger
    
    return -1

input_file = open('/content/sample_data/input 3.txt', 'r')
n, m = input_file.readline().split()
n, m = int(n), int(m)
graph = [[] for _ in range(n)]
for _ in range(m):
    line = input_file.readline().split()
    u, v, w = int(line[0]), int(line[1]), int(line[2])
    graph[u-1].append((v-1, w))
input_file.close()

ans = dijkstra(graph, 0, n-1)

output_file = open('/content/sample_data/output 3', 'w')
if ans == -1:
    output_file.write("Impossible\n")
else:
    output_file.write(str(ans) + '\n')
output_file.close()
```
