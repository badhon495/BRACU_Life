```python
file_in = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_8/Input_Output/input1.txt', 'r')
file_out = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_8/Input_Output/output1.txt', 'w')

cityNo, roadNo = map(int, file_in.readline().split())
roads = []
parent = list(range(cityNo + 1))


def creatingRoads(roads, roadNo):
    for i in range(roadNo):
        u, v, w = map(int, file_in.readline().split())
        roads.append((u, v, w))


def find(value):
        if parent[value] != value:
            parent[value] = find(parent[value])
        return parent[value]


def findMinCost(cityNo, roads):
    roads.sort(key=lambda x: x[2])
    min_cost = 0
    for u, v, w in roads:
        parentU = find(u)
        parentV = find(v)
        if parentU != parentV:
            parent[parentV] = parentU
            min_cost += w
    return min_cost


creatingRoads(roads, roadNo)
minCost = findMinCost(cityNo, roads)
file_out.write(str(minCost))
file_out.close()
```

```python
file_in = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_8/Input_Output/input2.txt', 'r')
file_out = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_8/Input_Output/output2.txt', 'w')
n = int(file_in.readline())

arr = [0 for i in range(n+2)]
def count(n):
    if arr[n]==0:
        if n<=2:
            arr[n] = 1
        else:
            arr[n] = count(n-1)+count(n-2)
    return arr[n]

file_out.write(str(count(n+1)))
file_out.close()
```

```python
file_in = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_8/Input_Output/input3.txt', 'r')
file_out = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_8/Input_Output/output3.txt', 'w')

coinAmount, target = map(int, file_in.readline().split(" "))
coinList = sorted(list(map(int, file_in.readline().split(" "))))


def coinCheck(target, coinList):
    arr = [target + 1] * (target+1)
    arr[0] = 0
    for i in range(1, target+1):
        for c in coinList:
            if i - c >= 0:
                arr[i] = min(arr[i], 1 + arr[i-c])
    return arr[target] if arr[target] != target+1 else -1


file_out.write(str(coinCheck(target, coinList)))
file_out.close()
```
