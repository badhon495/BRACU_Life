```python
file_in = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_1/Input_Output/input2.txt', 'r')
file_out = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_1/Input_Output/output2.txt', 'w')
len = int(file_in.readline())
values = [int(i) for i in file_in.readline().split(' ')]

def bubble_sort(list1, len):
  for i in range(len-1):
    swaped = False
    for j in range(len-1-i):
      if list1[j] > list1[j+1]:
        temp = list1[j]
        list1[j] = list1[j+1]
        list1[j+1] = temp
        swaped = True

    if not swaped:
      break

bubble_sort(values, len)

s = ''
for i in range(len):
    if i != (len-1):
        s+= str(values[i]) + ' '
    else:
       s+= str(values[i])
file_out.write(s)
file_out.close()


```

```python
file_in = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_1/Input_Output/input3.txt', 'r')
file_out = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_1/Input_Output/output3.txt', 'w')

loop = int(file_in.readline())
id = list(map(int,file_in.readline().split(' ')))
marks = list(map(int,file_in.readline().split(' ')))

def selection_sort(loop, id , marks):
  for i in range(loop):
    max_index = i

    for j in range(max_index+1, loop):
      if marks[j] == marks[max_index]:
        if id[j] < id[max_index]:  #ekhane mark same hoile, id check kortesi. id choto hoile swape kore ditesi. mark swape kortesi na cause korleo ki, na korleo ki
           id[j], id[max_index] = id[max_index], id[j]
      elif marks[j] > marks[max_index]:
        max_index = j

    if i != max_index:  #i and max_index same naki dekhtesi. same na thakle id and mark swape kore dibo.
      marks[i], marks[max_index] = marks[max_index], marks[i]
      id[i], id[max_index] = id[max_index], id[i]


selection_sort(loop, id, marks)

for i in range(loop-1):
  file_out.write(f'ID: {id[i]} Mark: {marks[i]}\n')
file_out.write(f'ID: {id[loop-1]} Mark: {marks[loop-1]}')
file_out.close()
```

```python
file_in = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_1/Input_Output/input4.txt', 'r')
file_out = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_1/Input_Output/output4.txt', 'w')

len = int(file_in.readline())
all_list = []

for i  in range(len): #this for loop only for taking proper input.
    read_line = file_in.readline().strip() #using strip to remove \n from last
    parts = read_line.split(" will departure for ")
    name = parts[0]
    loc = parts[1].split(" at ")[0]
    time = parts[1].split(" at ")[1]
    all_list.append([name, loc, time])

for i in range(len-1):  #doing bubble sort. cause bubble sort is a stable algo. in the last line of the question it asked us if it is a tie then first will come first
  for j in range(len-1-i):
    if all_list[j][0] > all_list[j+1][0]: #sorting for name.
      temp = all_list[j]
      all_list[j] = all_list[j+1]
      all_list[j+1] = temp
    elif (all_list[j][0]==all_list[j+1][0]) and (all_list[j][2]<all_list[j+1][2]): #name same, so sorting using time.
      temp = all_list[j]
      all_list[j] = all_list[j+1]
      all_list[j+1] = temp


for i in all_list[:len -1]:
  file_out.write(f'{(i[0])} will departure for {i[1]} at {i[2]}\n')
file_out.write(f'{(all_list[len-1][0])} will departure for {all_list[len-1][1]} at {all_list[len-1][2]}')

file_out.close()
```
