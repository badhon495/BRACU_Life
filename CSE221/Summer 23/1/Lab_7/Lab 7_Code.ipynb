```python
file_in = open("/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_7/Input_Output/input1.txt", "r")
file_out = open("/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_7/Input_Output/output1.txt", "w")

tasks = int(file_in.readline())
start_end = []
work = []


for i in range(tasks):
    start_end.append(tuple(map(int, file_in.readline().split(" "))))
start_end.sort(key = lambda x: x[1])

for i in start_end:
    if work == [] or work[-1][1] <= i[0]:
        work.append(i)


file_out.write(f'{str(len(work))}\n')
for i in work:
    file_out.write(f'{i[0]} {i[1]}\n')
file_out.close()
```

```python
file_in = open("/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_7/Input_Output/input2.txt", "r")
file_out = open("/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_7/Input_Output/output2.txt", "w")

taskNo, personNo = map(int, file_in.readline().split(' '))
time_range = []
personList = [[(0, 0)] for i in range(personNo)]
visited = []


for i in range(taskNo):
    time_range.append(tuple(map(int, file_in.readline().split(" "))))
time_range.sort(key=lambda x: x[1])

for time in time_range:
    for person in personList:
        if person == [(0, 0)] or person[-1][-1] <= time[0]:
            person.append(time)
            visited.append(time)
            personList.sort(key=lambda x: x[-1][-1], reverse=True)
            break


file_out.write(str(len(visited)))
file_out.close()
```

```python
file_in = open("/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_7/Input_Output/input3.txt", "r")
file_out = open("/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_7/Input_Output/output3.txt", "w")

peopleNo, quriesNo = map(int, file_in.readline().split())
connections = []
connectedList = [i for i in range(peopleNo+1)]

for i in range(quriesNo):
    connections.append(tuple(map(int, file_in.readline().split(" "))))


def total_amount(value, connectedList):
    count = 0
    for i in connectedList:
        if i == value:
            count += 1
    return count


def change(fromValue, toValue, connectedList):
    for i in range(len(connectedList)):
        if connectedList[i] == fromValue:
            connectedList[i] = toValue
    return connectedList


def union(connections, connectedList):
    changed = None
    for i in connections:
        u, v = i

        if connectedList[u] == connectedList[v]:
            file_out.write(f'{(total_amount(connectedList[u], connectedList))}\n')
            continue

        elif connectedList[u] != u and connectedList[v] != v:
            if total_amount(connectedList[u], connectedList) > total_amount(connectedList[v], connectedList):
                connectedList = change(
                    connectedList[v], connectedList[u], connectedList)
                changed = connectedList[u]

            elif total_amount(connectedList[u], connectedList) < total_amount(connectedList[v], connectedList):
                connectedList = change(
                    connectedList[u], connectedList[v], connectedList)
                changed = connectedList[v]

        elif connectedList[u] == u or connectedList[v] == v:
            if connectedList[u] == u and connectedList[v] == v:
                connectedList = change(
                    connectedList[v], connectedList[u], connectedList)
                changed = connectedList[u]

            elif connectedList[u] != u and connectedList[v] == v:
                connectedList = change(
                    connectedList[v], connectedList[u], connectedList)
                changed = connectedList[u]

            elif connectedList[u] == u and connectedList[v] != v:
                connectedList = change(
                    connectedList[u], connectedList[v], connectedList)
                changed = connectedList[v]
        file_out.write(f'{(total_amount(changed, connectedList))}\n')


union(connections, connectedList)
file_out.close()
```
