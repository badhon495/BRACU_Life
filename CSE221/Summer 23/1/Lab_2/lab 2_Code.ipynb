```python
file_in = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_2/Input_Output/input2.txt', 'r')
file_out = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_2/Input_Output/output2_2.txt', 'w')

len_one = int(file_in.readline())
int_list_one = list(map(int, file_in.readline().split(' ')))

len_two = int(file_in.readline())
int_list_two = list(map(int, file_in.readline().split(' ')))

sorted_list = []
i, j = 0,0



while j != (len_two)  and i != (len_one): # (len-1) porjnto cholbe.
    if int_list_one[i] > int_list_two[j]:
        sorted_list.append(int_list_two[j])
        j+=1
    else:
        sorted_list.append(int_list_one[i])
        i+=1

# duita list shoman na hole tahole baki ja thakbe eikhan theke append hobe. and as it is a sorted list so this values will be equal or greater.
if i != len_one:
    for k in range(i, len_one):
        sorted_list.append(int_list_one[k])
if j != len_two:
    for k in range(j, len_two):
        sorted_list.append(int_list_two[k])


for i in sorted_list:
    file_out.write(f'{i} ')

file_out.close()

#this thing is baically O(n+k), but the k loop in the most of the case is smaller, so it will be ignored.
```

```python
file_in = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_2/Input_Output/input3.txt', 'r')
file_out = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_2/Input_Output/output3.txt', 'w')

length = int(file_in.readline())
arr1 = list(map(int, file_in.readline().split(' ')))



def merge(a1, a2):
    new_arr = []
    i = j = 0
    while i < len(a1) and j < len(a2): #ekhane shb venge venge check hobe until i or j kono array er length er shoman hoi
        if a1[i] <= a2[j]:
            new_arr.append(a1[i])
            i += 1
        else:
            new_arr.append(a2[j])
            j += 1

#i or j shoman hoye jawa mane ekta array sesh, arekta array baki. oi baki part jehetu same array er tai already sorted. ekhane just baki part tuku add kore dicchi
    while i < len(a1):
            new_arr.append(a1[i])
            i += 1

    while j < len(a2):
            new_arr.append(a2[j])
            j += 1

    return new_arr


def mergeSort(arr1):
    if len(arr1) <= 1:
        return arr1
    else:
        mid = len(arr1)//2
    a1 = mergeSort(arr1[:mid])
    a2 = mergeSort(arr1[mid:])
    return merge(a1, a2)

sorted_list = mergeSort(arr1)

for i in sorted_list:
    file_out.write(f'{i} ')

file_out.close()

#merge sort so nlogn
```
