```python
from heapq import heappop, heappush
file_in = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_6/Input_Output/input1.txt', 'r')
file_out = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_6/Input_Output/output1.txt', 'w')

vertices, edges = list(map(int, file_in.readline().split(' ')))
path = file_in.read().split('\n')
graph = {}
start = int(path.pop(-1))


def creating_graph(graph, path):
    for i in path:
        u, v, w = list(map(int, i.split(' ')))
        if u not in graph:
            graph[u] = [(v,w)]
        else:
            graph[u].append((v,w))


def dijkstra(graph, start):
    vis_value = [float('inf') for i in range(vertices+1)]
    minHeap = [(start, 0)]
    visited = set()

    while minHeap:
        vertice, weight = heappop(minHeap)
        if vis_value[vertice] > weight:
            vis_value[vertice] = weight

        if vertice in visited:
            continue

        visited.add(vertice)
        if vertice in graph:
            for newV, newW in graph[vertice]:
                heappush(minHeap, (newV, newW+weight))

    return vis_value


creating_graph(graph,path)
short_path = [-1 if type(i) != int else i for i in dijkstra(graph, start)]
file_out.write(' '.join(map(str, short_path[1:])))
file_out.close()
```
