```python
file_in = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_3/Input_Output/input2.txt', 'r')
file_out = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_3/Input_Output/output2.txt', 'w')
size = int(file_in.readline())
arr1 = list(map(int,file_in.read().split()))


def get_cross_max(arr): #array te 2 er beshi value hoilei cross e pathabo.
    left = float("-inf")
    right = float("-inf")

    for i in range(0, len(arr)//2):
        if arr[i] > left: #ekhane asholei boro value ta neoa hoitese.
            left = arr[i]

    for i in range(len(arr)//2,len(arr)):
        if abs(arr[i]) > right: #ekhane abs kortesi cause j hobe i er right side e thaka array gular vetorer kono value and j sqare kora hoitese. so negative does not matter.
            right = abs(arr[i])
    return left+right**2


def get_max(arr):
    if len(arr) == 1:
        return float("-inf") #negative infinity dhortesi cause 1 length er array diye kono din max value pawa jai na.
    if len(arr) == 2:
        return arr[0] + arr[1]**2
    left_max = get_max(arr[:len(arr)//2])
    right_max = get_max(arr[len(arr)//2:])
    cross_max = get_cross_max(arr)
    return max(left_max, right_max,cross_max)


file_out.write(str(get_max(arr1)))
file_out.close()
```

```python
file_in = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_3/Input_Output/input3.txt', 'r')
file_out = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_3/Input_Output/output3.txt', 'w')
size = int(file_in.readline())
arr1 = list(map(int,file_in.read().split()))

def parting(arr, start, end):
    pivot = arr[end]
    idx = start-1

    for i in range(start, end):
        if arr[i] < pivot:
            idx += 1
            arr[i], arr[idx] = arr[idx], arr[i]
    arr[idx+1], arr[end] = arr[end], arr[idx+1]
    return idx+1


def quickSort(arr, start, end):
    if start >= end:
        return

    part = parting(arr, start, end)
    quickSort(arr, start, part-1)
    quickSort(arr, part+1, end)


quickSort(arr1, 0, size-1)
file_out.write(" ".join(map(str, arr1)))
file_out.close()
```

```python
file_in= open("/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_3/Input_Output/input4.txt","r")
file_out= open("/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_3/Input_Output/output4.txt","w")

size= int(file_in.readline())
arr1= list(map(int,file_in.readline().split(" ")))
querie_amount = int(file_in.readline())
querie_index = list(map(int,file_in.read().split("\n")))


def partition(arr, left, right):
    pivot = arr[left]
    i = left + 1

    for j in range(left+1, right+1):
        if arr[j] < pivot:
            arr[i], arr[j] = arr[j], arr[i]
            i += 1

    arr[i - 1], arr[left] = arr[left], arr[i - 1]
    return i - 1



def kthSmallest(arr, left, right, k):
    if left == right:
        return arr[left]
    pivotIndex = partition(arr, left, right)

    if k == pivotIndex:
        return arr[pivotIndex]
    elif k < pivotIndex:
        return kthSmallest(arr, left, pivotIndex - 1, k)
    return kthSmallest(arr, pivotIndex + 1, right, k)


values = [kthSmallest(arr1, 0, size - 1, querie_index[i]-1) for i in range(querie_amount)]
file_out.write("\n".join(map(str, values)))
file_out.close()
```
