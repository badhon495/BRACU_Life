```python
# q1
file_in = open("/content/drive/MyDrive/Course Material/CSE221/Question/Lab Final/Input_Output/input.txt", 'r')
file_out = open('/content/drive/MyDrive/Course Material/CSE221/Question/Lab Final/Input_Output/output_A.txt', 'w')

nodeCount, edgeCount = map(int, file_in.readline().split(' '))
graph = {}
visited = []


# q2
def graphBuild(graph, edgeCount):
    for i in range(edgeCount):
        u, v = map(int, file_in.readline().split(' '))

        if u in graph:
            graph[u].append(v)
        else:
            graph[u] = [v]

# q3
def BFS(graph, start, visited):
    queue = [start]

    while queue:
        node = queue.pop(0)
        visited.append(node)
        if node in graph:
            for i in graph[node]:
                queue.append(i)


# q4
def countLevel(graph, start, count_list):
    queue = [start]
    while queue:
        node = queue.pop(0)
        if node in graph:
            for i in graph[node]:
                count_list[i] = count_list[node]+1
                queue.append(i)



graphBuild(graph, edgeCount)
start = int(file_in.readline())

BFS(graph, start, visited)

count = 0
count_list = [0 for i in range(nodeCount+1)]
countLevel(graph, start, count_list)

# q5
file_out.write(f'the path from source {start} each node levelwise:\n')
file_out.write(str(visited)[1:-1].replace(',', ' '))
file_out.write(f'\nin order to reach {nodeCount} user has to move {str(count_list[-1])} floors from the source node {start}')
file_out.close()
file_in.close()
```
