```python
file_in = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_4/Input_Output/Task 05/input5_3.txt', 'r')
file_out = open('/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_4/Input_Output/Task 05/output5.txt', 'w')

cities, roads, destination = list(map(int,file_in.readline().split(' ')))
path = file_in.read().split('\n')
graph = {}
#adding 1 because in this question dora always begins her journy from 1. so it will always be our first node to traverse.
queue = [[1]]
visited = []

#created graph
def creating_graph(graph, path):
    for i in path:
        u, v = list(map(int, i.split(' ')))

        if u not in graph:
            graph[u] = [v]
        else:
            graph[u].append(v)

        if v not in graph:
            graph[v] = [u]
        else:
            graph[v].append(u)


def BFS(graph, queue, visited, destination):
#checking if our goal is 1 or not. if destination is 1 then no need do other bullshit
    start = queue[0]
    if start[0] == destination:
        return start

#queue vortesi ar check kortesi
    while queue:
        path = queue.pop(0)
#doing -1 cause path adding like this [first node, then the node a called using first node, then second node if call another node that will be added here]
#so we are using last node to call another node
        node = path[-1]

        if node not in visited:
            neighbours = graph[node]
#here traversing the node and its connected node. if matchs with destination then retun the new path
            for neighbour in neighbours:
                new_path = list(path)
                new_path.append(neighbour)
                queue.append(new_path)
                if neighbour == destination:
                    return new_path
#adding it to visited. otherwise for cycle it will run infinite loop
            visited.append(node)

creating_graph(graph, path)
steps = (BFS(graph, queue, visited, destination))
file_out.write(f'Time: {len(steps)-1}\nShortest Path: {" ".join(map(str,steps))}')
file_out.close()
```

```python
file_in = open("/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_4/Input_Output/Task 06/input6_3.txt","r")
file_out = open("/content/drive/MyDrive/Course Material/CSE221/Lab Assignment/Lab_4/Input_Output/Task 06/output6.txt","w")
R,H = [int(i) for i in file_in.readline().split(" ")]

#Initializing the grid
grid = [[] for i in range(R) ]

#Inserting the input elements into the grid
for i in range(R):
  for j in file_in.readline().split()[0] :
    grid[i].append(j)

def floodFill(r,c,g):
  #Checking whether the rows or columns exceed boundaries or the grid has any "#"
  if c < 0 or r < 0 or r>= R or c >= H or grid[r][c] == "#":
    return 0

  if grid[r][c] == "D":
    diamond = 1
  else:
    diamond = 0

  grid[r][c] = "#"
  diamond+=floodFill(r-1,c,g) #Going Up
  diamond+=floodFill(r+1,c,g) #Going Down
  diamond+=floodFill(r,c-1,g) #Going Left
  diamond+=floodFill(r,c+1,g) #Going Right

  return diamond

max_diamond = 0 #Takes the count of maximum diamonds
for line in range(R):
  for char in range(H):
    if grid[line][char] == ".":
      diamonds = floodFill(line,char,grid)
      max_diamond = max(max_diamond,diamonds)

file_out.write(f'{max_diamond}')
file_out.close()
```
